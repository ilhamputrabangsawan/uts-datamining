### Soal pre-processing data

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
````

````python
startup_data = pd.read_csv('50_Startups.csv')
````

````python
startup_data.sample(7)

# Analisis data kosong
missing_values = startup_data.isnull().sum()
print("Data kosong:\n", missing_values)

# Mengisi data kosong dengan nilai mean
mean_imputer = SimpleImputer(strategy='mean')
startup_data[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']] = mean_imputer.fit_transform(startup_data[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']])

# One-hot encoding untuk kolom 'State'
encoder = OneHotEncoder()
state_encoded_matrix = encoder.fit_transform(startup_data[['State']])
state_encoded_df = pd.DataFrame(state_encoded_matrix.toarray(), columns=encoder.get_feature_names_out(['State']))
startup_data = pd.concat([startup_data, state_encoded_df], axis=1)
startup_data.drop('State', axis=1, inplace=True)

# Membuat kolom baru 'Tax'
startup_data['Tax'] = (startup_data['Profit'] + startup_data['Marketing Spend'] + startup_data['Administration']) * 0.05

# StandardScaler untuk semua fitur kecuali 'Profit'
scaler = StandardScaler()
columns_to_scale = [col for col in startup_data.columns if col != 'Profit']
startup_data[columns_to_scale] = scaler.fit_transform(startup_data[columns_to_scale])

# Menampilkan data setelah preprocessing
print("Data setelah preprocessing:\n", startup_data.head())

````

### Soal
Dari data di atas lakukan pre-processing sbb:
1. Lakukan analisa mengenai field mana saja yang memiliki data kosong(Nan) lalu Isi data yang kosong tersebut menggunakan nilai mean
2. lakukan oneHotEncoder ke field state
3. buat field baru yaitu Tax. dimana Tax = (profit + Marketing Spend + administration) * 5%
4. Jalankan standardScaler field-field tersebut
